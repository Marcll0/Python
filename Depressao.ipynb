```python
dataset = pd.get_dummies(dataset, columns=['Sleep Duration'], prefix='', prefix_sep='')
dataset = pd.get_dummies(dataset, columns=['Dietary Habits'], prefix='', prefix_sep='')
dataset.head()
```

```python
for i in range(len(dataset)):
  if dataset['Gender'][i] == "Male" :
    dataset.loc[i,"Gender"] = 1
  else:
    dataset.loc[i,"Gender"] = 0

  if dataset['Have you ever had suicidal thoughts ?'][i] == "Yes" :
    dataset.loc[i,"Have you ever had suicidal thoughts ?"] = 1
  else:
    dataset.loc[i,"Have you ever had suicidal thoughts ?"] = 0

  if dataset['Family History of Mental Illness'][i] == "Yes" :
    dataset.loc[i,"Family History of Mental Illness"] = 1
  else:
    dataset.loc[i,"Family History of Mental Illness"] = 0

  if dataset['Depression'][i] == "Yes" :
    dataset.loc[i,"Depression"] = 1
  else:
    dataset.loc[i,"Depression"] = 0

  if dataset['5-6 hours'][i] == True :
    dataset.loc[i,"5-6 hours"] = 1
  else:
    dataset.loc[i,"5-6 hours"] = 0

  if dataset['7-8 hours'][i] == True :
    dataset.loc[i,"7-8 hours"] = 1
  else:
    dataset.loc[i,"7-8 hours"] = 0

  if dataset['Less than 5 hours'][i] == True :
    dataset.loc[i,"Less than 5 hours"] = 1
  else:
    dataset.loc[i,"Less than 5 hours"] = 0

  if dataset['More than 8 hours'][i] == True :
    dataset.loc[i,"More than 8 hours"] = 1
  else:
    dataset.loc[i,"More than 8 hours"] = 0

  if dataset['Healthy'][i] == True :
    dataset.loc[i,"Healthy"] = 1
  else:
    dataset.loc[i,"Healthy"] = 0

  if dataset['Moderate'][i] == True :
    dataset.loc[i,"Moderate"] = 1
  else:
    dataset.loc[i,"Moderate"] = 0

  if dataset['Unhealthy'][i] == True :
    dataset.loc[i,"Unhealthy"] = 1
  else:
    dataset.loc[i,"Unhealthy"] = 0
```

# Modelos de Classificação

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = dataset.loc[:, ['Gender', 'Age', 'Academic Pressure', 'Study Satisfaction',
       'Have you ever had suicidal thoughts ?', 'Study Hours',
       'Financial Stress', 'Family History of Mental Illness',
       '5-6 hours', '7-8 hours', 'Less than 5 hours', 'More than 8 hours',
       'Healthy', 'Moderate', 'Unhealthy']].values
y = dataset.loc[:, 'Depression'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
y_train = y_train.astype(int)
y_test = y_test.astype(int)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

classifier1 = LogisticRegression(class_weight='balanced',random_state = 0)
classifier1.fit(X_train, y_train)


y_pred1 = classifier1.predict(X_test)
cm = confusion_matrix(y_test, y_pred1)
print(cm)
accuracy_score(y_test, y_pred1)
```

```python
print(np.bincount(y_train))
print(np.bincount(y_test))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(classifier1, X_train, y_train, cv=5)
print("Acurácia média (cross-validação):", scores.mean())
```

```python
# Coeficientes do modelo
coefficients = classifier1.coef_[0]

# Nome das variáveis
feature_names = ['Gender', 'Age', 'Academic Pressure', 'Study Satisfaction',
                 'Have you ever had suicidal thoughts ?', 'Study Hours',
                 'Financial Stress', 'Family History of Mental Illness',
                 '5-6 hours', '7-8 hours', 'Less than 5 hours', 'More than 8 hours',
                 'Healthy', 'Moderate', 'Unhealthy']

# DataFrame para visualização
coef_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})
print(coef_df.sort_values(by='Coefficient', ascending=False))
```
